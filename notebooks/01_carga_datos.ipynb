## 2. Carga de datos — Matrícula escolar

Fuente: Ministerio de Educación de la Nación  
Archivos: matrícula total y matrícula por edad, años 2022–2024  
Formato: CSV con separador `;` y encoding `latin-1`

```python
# Matrícula total por departamento — 2022, 2023, 2024
df_mat_2022 = pd.read_csv("../data/raw/2022 Matricula - agregada(in).csv", sep=";", encoding="latin-1")
df_mat_2023 = pd.read_csv("../data/raw/2023 Matricula - agregada(in).csv", sep=";", encoding="latin-1")
df_mat_2024 = pd.read_csv("../data/raw/2024 Matricula - agregada(in).csv", sep=";", encoding="latin-1")

# Matrícula por edad — 2022, 2023, 2024
df_edad_2022 = pd.read_csv("../data/raw/2022 Matricula por edad - agregada(in).csv", sep=";", encoding="latin-1")
df_edad_2023 = pd.read_csv("../data/raw/2023 Matricula por edad - agregada(in).csv", sep=";", encoding="latin-1")
df_edad_2024 = pd.read_csv("../data/raw/2024 Matricula por edad - agregada(in).csv", sep=";", encoding="latin-1")

print("=== MATRÍCULA TOTAL ===")
print("2022:", df_mat_2022.shape)
print("2023:", df_mat_2023.shape)
print("2024:", df_mat_2024.shape)

print("\n=== MATRÍCULA POR EDAD ===")
print("2022:", df_edad_2022.shape)
print("2023:", df_edad_2023.shape)
print("2024:", df_edad_2024.shape)
```

## 3. Carga de datos — Abandono interanual

Fuente: Ministerio de Educación de la Nación  
Archivos: tasas de abandono interanual, dos series históricas  
- `2024-2012`: serie reciente por división político-territorial  
- `2016-2003`: serie histórica con estructura nacional homogénea  

**Nota técnica:** los archivos Excel tienen encabezado institucional en las primeras 10 filas.  
Se usa `skiprows=10` para saltar esas filas y leer directamente los datos.

```python
# Salteamos las primeras 10 filas que son encabezado institucional
abandono_reciente = pd.read_excel(
    "../data/raw/Tasa de Abandono Interanual 2024-2012 según división político-territorial.xlsx",
    skiprows=10
)

abandono_historico = pd.read_excel(
    "../data/raw/Tasa de Abandono Interanual 2016-2003 según estructura nacional homogenea.xlsx",
    skiprows=10
)

print("Reciente:", abandono_reciente.shape)
print("Histórico:", abandono_historico.shape)
abandono_reciente.head(5)
```
